```python
from pymongo import MongoClient

import pandas as pd
import numpy as np
```

```python
database  = MongoClient()["scraper2"]
movements = database["container_movements"]
statuses  = database["statuses"]
```

```python
select = {
    "_id": 0,
    "container": 1,
    "date": 1,
    "location": 1,
    "latitude": 1,
    "longitude": 1,
    "status": 1,
    "estimated": 1,
    "transport_type": 1
}
```

## Maersk

```python
maersk_query = {
    "container": "MRKU5264777"
}

maersk = []
cursor = movements.find(maersk_query, select).sort("date", 1)

for movement in cursor:
    # Vehicle type as number
    movement["vehicle"] = 1 if movement["transport_type"] == "Vessel" else 0
    # Status type as number
    status = statuses.find_one({"Maersk": movement["status"]})
    movement["status_code"] = int(status["code"]) if status else 0
    # Append to list
    maersk.append(movement)

print(len(maersk))
```

```python
maersk_df = pd.DataFrame(maersk)
maersk_df
```
